#### Chart - 4

````python
# Chart - 4 visualization code
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('Hotel Bookings.csv')

# Aggregate the number of bookings by country
country_bookings = df['country'].value_counts().reset_index()
country_bookings.columns = ['country', 'bookings']

top_n = 10
top_countries = country_bookings.head(top_n)['country']
top_country_data = df[df['country'].isin(top_countries)]

# Aggregate the data by agent and country for the top N countries
agent_country_data = top_country_data.groupby(['agent', 'country']).size().reset_index(name='bookings')

# Pivot the data to get agents as columns and countries as rows
pivot_data = agent_country_data.pivot(index='country', columns='agent', values='bookings').fillna(0)

# Plotting stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(15, 8))

plt.title(f'Number of Bookings by Agent for Top {top_n} Countries')
plt.xlabel('Country')
plt.ylabel('Number of Bookings')
plt.legend(title='Agent')
plt.tight_layout()
plt.show()



````

##### 1. Why did you pick the specific chart?

This chart makes the visualisaton clear and simplified for us to understand the direct relation between number of bookings by different agents for the top 10 countries.

##### 2. What is/are the insight(s) found from the chart?

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('Hotel Bookings.csv')

# Ensure that the arrival_date_year, arrival_date_month, and arrival_date_day_of_month columns are in the correct format
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' +
                                    df['arrival_date_month'] + '-' +
                                    df['arrival_date_day_of_month'].astype(str), errors='coerce')

# Create a new column for the day of the week
df['day_of_week'] = df['arrival_date'].dt.day_name()

# Filter data for weekend nights (Saturday and Sunday)
weekend_data = df[df['day_of_week'].isin(['Saturday', 'Sunday'])]

# Aggregate the number of stays by the number of adults
adults_stays = weekend_data.groupby('adults').size().reset_index(name='stays')

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(adults_stays['adults'], adults_stays['stays'])

# Customize the plot
plt.title('Number of Stays on Weekend Nights by Number of Adults')
plt.xlabel('Number of Adults')
plt.ylabel('Number of Stays')

# Show the plot
plt.show()

````
